```python
import os
path = '/content/'
extension = '.csv'
```

```python
files = [file for file in os.listdir(path) if file.endswith(extension)]
```

```python
dfs = []
for file in files:
    dataset = pd.read_csv(os.path.join(path, file))
    dfs.append(dataset)
```

```python
dataset = pd.concat(dfs, ignore_index=True)
display(dataset)
```

```python
dataset = dataset.drop(columns=['headline'])
display(dataset)
```

```python
dataset.drop_duplicates(inplace=True)
display(dataset)
```

```python
output_file_name = 'datasetmerged.csv'

# Save the DataFrame to a CSV file
dataset.to_csv(output_file_name, index=False)
```
